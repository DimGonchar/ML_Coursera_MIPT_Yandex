```python
from sklearn.datasets import load_digits
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = load_digits()
```

```python
data
```

```python
X = data.data
Y = data.target
```

```python
Y
```

```python
X.shape
```

```python
set(list(Y))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.35,  shuffle=False)
```

```python
def euclidean_distance(vector_1, vector_2):
    return np.sqrt(np.sum((vector_1-vector_2)**2))
```

```python
def one_NN(X_train, X_test, Y_train, Y_test):
    Y_pred = []
    for vector in X_test:
        list_information = [0, euclidean_distance(vector, X_train[0])]
        
        for index_vector_train in range(1, len(X_train)):
            metric = euclidean_distance(vector, X_train[index_vector_train])
            
            if metric < list_information[1]:
                list_information[0] = index_vector_train
                list_information[1] = metric
         
        Y_pred.append(Y_train[list_information[0]])
    return Y_pred
```

```python
%%time
Y_pred = one_NN(X_train, X_test, Y_train, Y_test)
```

```python
Y_pred
```

```python
result_oneNN = accuracy_score(Y_test, Y_pred)
```

```python
result_oneNN
```

```python
result_oneNN = 1 - result_oneNN
```
