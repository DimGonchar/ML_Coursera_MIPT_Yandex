```python
import numpy as np
from scipy.optimize import minimize
from scipy.optimize import differential_evolution
from matplotlib import pylab as plt
```

```python
def my_func(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
```

```python
x = np.arange(1, 30, 0.2)
y = my_func(x)
```

```python
plt.plot(x, y)
plt.show()
```

# Задача 1

```python
for initial_approximation in range(30):
    result_1 = minimize(my_func, np.array([initial_approximation]),  method='Nelder-Mead')
    print("_________________---------------------_____________________")
    print("initial_approximation: ", initial_approximation)
    print(result_1)
    
```

```python
for initial_approximation in range(30):
    result2 = minimize(my_func, np.array([initial_approximation]), method = "BFGS")
    print("_____________________--------------------------_____________________-")
    print("initial_approximation: ", initial_approximation)
    print(result2)
```

```python
answer1 = str(round(minimize(my_func, [2], method = "BFGS").fun, 2)) + " " + \
str(round(minimize(my_func, [30], method = "BFGS").fun, 2))

```

```python
with open("answer_one", "w") as f:
    f.write(answer1)
```

```python
with open("answer_one", "r") as f:
    print(f.read())
```

```python
def new_my_func(x):
    if type(x) == float or type(x) == int:
        return int(my_func(x))
    else:
        return np.array([int(my_func(elem)) for elem in x])
    
```

```python
y2 = new_my_func(x)
```

```python
plt.plot(x, y2)
plt.show()
```
