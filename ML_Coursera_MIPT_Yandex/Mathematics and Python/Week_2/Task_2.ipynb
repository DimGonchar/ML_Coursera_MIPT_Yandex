```python
import numpy as np
from scipy import linalg
from matplotlib import pylab as plt
```

```python
def my_func(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
```

```python
def search_polynom(list_point, function):
    a = [[point**index for index in range(len(list_point))] for point in list_point]
    b = [my_func(point) for point in list_point]
    w_i = linalg.solve(a, b)
    
    return w_i
    
```

 Многочлен первой степени

```python
result1 = search_polynom([1, 15], my_func)
```

```python
result1
```

Второй

```python
result2 = search_polynom([1, 8, 15], my_func)
```

```python
result2
```

Третьей

```python
result3 = search_polynom([1, 4, 10, 15], my_func)
```

```python
result3
```

```python
def function_plot(list_koef, x):
    result = 0
    for index_elem in range(len(list_koef)):
        result += list_koef[index_elem] * x**index_elem
    return result
```

```python
x = np.arange(1, 15, 0.2)
y = my_func(x)
plt.plot(x, y, 'r')
plt.plot(x, function_plot(result1, x))
plt.plot(x, function_plot(result2, x))
plt.plot(x, function_plot(result3, x))
```
