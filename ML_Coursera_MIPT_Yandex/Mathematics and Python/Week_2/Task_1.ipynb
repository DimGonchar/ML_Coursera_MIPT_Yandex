```python
import re
import numpy as np
from scipy.spatial import distance
```

```python
filename = "sentences.txt"
with open(filename, "r") as f:
     list_line = f.readlines()
list_line = [line.rstrip() for line in list_line]
```

```python
list_line
```

```python
list_line = [line.lower() for line in list_line]
```

```python
list_line
```

```python
list_line = [re.split('[^a-z]', line) for line in list_line]
for i in range(len(list_line)):
    list_line[i] = [x for x in  list_line[i] if x]
```

```python
list_line
```

```python
set_words = set(list_line[0])
for i in range(1, len(list_line)):
    set_words = set_words | set(list_line[i])
    
```

```python
set_words
```

```python
dict_words = {}
list_words = list(set_words)
for i in range(len(list_words)):
    dict_words[list_words[i]] = i
    
```

```python
dict_words
```

```python
matrix = np.eye(len(list_line), len(set_words))
```

```python
matrix.shape
```

```python
list_line
```

```python
for elem in range(len(list_line)):
    for word in list_line[elem]:
        matrix[elem][dict_words[word]] = list_line[elem].count(word)
```

```python
matrix
```

```python
result_list = []
for i in range(1, len(matrix)):
    result_list.append(distance.cosine(matrix[0], matrix[i]))
```

```python
result_list
```

```python
sort_result_list = sorted(result_list)
```

```python
sort_result_list
```

```python
result_str = str(result_list.index(sort_result_list[0])+1) + " " + str(result_list.index(sort_result_list[1])+1)
```

```python
result_str
```

```python
with open ("result_one", "w") as f:
    f.write(result_str)
```
